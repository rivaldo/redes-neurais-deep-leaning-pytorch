```python

import numpy as np
import matplotlib.pyplot as plt
a = -1
b = 4
c = 0.4

# ax + by + c = 0 EQUAÇÃO DA RETA
# y = (-a*x - c)/b
def plotline(a, b, c):
    x = np.linspace(-2, 4, 50)
    y = (-a*x - c)/b
    plt.figure(figsize=(14,10))  #largura X altura
    plt.axvline(0, -1, 1, color = 'k', linewidth = 1)
    plt.axhline(0, -2, 4, color = 'k', linewidth = 1)
    plt.grid(True)
    plt.plot(x,y)
plotline(a, b, c)
```

```python
p1 = ( 2, 0.4)
p2 = (1, 0.6)
p3 = (3, -0.4)

retorno1 = a * p1[0] + b * p1[1] + c
retorno2 = a * p2[0] + b * p2[1] + c
retorno3 = a * p3[0] + b * p3[1] + c

print(f'Resultado da equação para os ponto x = 2, y = 0.4 tem que ser zero, pois satisfaz a equação da reta ax + by + c = 0: {retorno1: .2f}')
print(f'Resultado da equação para os ponto x = 1, y = 0.6 tem que ser zero, pois satisfaz a equação da reta ax + by + c = 0: {retorno2: .2f}')
print(f'Resultado da equação para os ponto x = 3, y = -0.4 tem que ser zero, pois satisfaz a equação da reta ax + by + c = 0: {retorno3: .2f}')
```

*Os retornos positivos informam que os pontos estão acima da reta. Os retornos negativos, informam que os pontos estão abaixo da reta.*

 - f(x) = 0, define pontos na reta;
 - f(x) > 0, define pontos acima da reta;
 - f(x) < 0, define pontos abaixo da reta.

```python
plotline(a,b,c)
plt.plot(p1[0], p1[1], color = 'g', marker = 'o')
plt.plot(p2[0], p2[1], color = 'r', marker = 'o')
plt.plot(p3[0], p3[1], color = 'b', marker = 'o')
```

```python
from sklearn.datasets import make_classification

np.random.seed(46)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)

print(f'X: {X.shape}, Y: {Y.shape}')
```

**Do lado azul, temos as classes que valem 0, Y = 0. Dol lado amarelo, as classes que valem 1, Y = 1.**

```python
def plotmodel(w1, w2, b):
    
    
    
    x = np.linspace(-2, 4, 50)
    y = (-w1*x - b)/w2
    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax) 
    print(ymin, ymax)
    
   
    plt.figure(figsize=(14,10))  #largura X altura
    
    plt.axvline(0, -1, 1, color = 'k', linewidth = 1)
    plt.axhline(0, -2, 4, color = 'k', linewidth = 1)
    
    plt.scatter(X[:,0], X[:,1], marker='o', c=Y, edgecolors='k',)
    plt.plot(x,y)

    plt.grid(True)
    
    
    
    
w1 = 5 #a
w2 = 1 #b
b = 0.4 #c
plotmodel(w1, w2, b)
```

```python
def classify (ponto, w1, w2, b):
    retorno = w1 * ponto[0] + w2 * ponto[1] + b
    
    if retorno >= 0:
        return 1, 'yellow'
    else:
        return 0, 'blue'
    
p= (2, -1)
classe, cor = classify(p, w1, w2, b)
print(f'CLASSE: {classe} - COR: {cor}')
plotmodel(w1, w2, b)
plt.plot(p[0],p[1], marker='^', color=cor, markersize=40)
```

```python
acertos = 0
for k in range(len(X)):
    categ, _ = classify(X[k], w1, w2, b)
    if categ == Y[k]:
        acertos+=1
        
        
print(f'Acurácia: {acertos/len(X):.2%}')
```
